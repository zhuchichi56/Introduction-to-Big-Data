```python
import pandas as pd
credit =pd.read_csv("credit.csv")
credit.info()
credit

#这里显示object数据就是定性的
```

```python
#需要做数据预处理
to_process = [0,2,3,5,6,8,9,11,13,14,16,18,19]
columns = credit.columns[to_process]
#得到列的名字;
credit[columns]
```

```python

sub =credit['checking_balance'].factorize()
sub_tag = sub[1]
sub_tag
label = dict(zip(sub_tag,range(len(sub_tag))))
label
```

```python
#接下来是是将多列都变成数值型形式：
```

```python
labels = {}

for column in columns:
    data,label = credit[column].factorize()
    credit[column] = data
    labels[column] = dict(zip(label,range((len(label)))))


credit
#object数据onehotencoding会好很多？？
#整理标签总结了


```

```python
#开始训练：
from sklearn.model_selection import train_test_split
X = credit.iloc[:,:-1]
y = credit.default
x_train,x_test ,y_trian,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
credit_model = DecisionTreeClassifier(criterion = 'gini')
credit_model = DecisionTreeClassifier(criterion = 'entropy')
credit_model.fit(x_train,y_trian)
credit_model.score(x_test,y_test)
```

```python
#图形化一个树：
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(50,50))
#filled=True 把分支的label给放进去
plot_tree(credit_model,filled=True,fontsize=8)
plt.title('entropy')
plt.show()
```

```python


credit_model = DecisionTreeClassifier(criterion = 'gini')
credit_model.fit(x_train,y_trian)
credit_model.score(x_test,y_test)
#gini指数
from sklearn.tree import plot_tree
plt.figure(figsize=(50,50))
plot_tree(credit_model,filled=True,fontsize=8)
plt.title('gini')
plt.show()
```

```python

#决策树的剪去枝叶;

#前减枝
credit_model = DecisionTreeClassifier(criterion = 'gini',max_depth=5)
credit_model.fit(x_train,y_trian)
credit_model.score(x_test,y_test)
#gini指数
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
plot_tree(credit_model,filled=True,fontsize=8)
plt.title('gini')
plt.show()
```

```python
#post

dtree_mode  = DecisionTreeClassifier(criterion='gini')
path = dtree_mode.cost_complexity_pruning_path(x_train,y_trian)
path.ccp_alphas



path.impurities
```

```python
#alpha 和不纯度之间的关系；
plt.plot(path.ccp_alphas,path.impurities,marker = 'o')
```

```python
model_trees =[]
for i in path.ccp_alphas:
    model = DecisionTreeClassifier(criterion='gini',ccp_alpha=i)
    model.fit(x_train,y_trian)
    model_trees.append(model.score(x_test,y_test))
    print(model.tree_.node_count)
    print(model.tree_.max_depth)


best =path.ccp_alphas[model_trees.index(max(model_trees))]
```

```python
plt.plot(path.ccp_alphas,model_trees)
best




```
