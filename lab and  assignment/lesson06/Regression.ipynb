```python

import numpy as np
import pandas as pd
import sklearn.datasets as datasets
```

```python
boston =datasets.load_boston()
boston.feature_names
bos = pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
boston.feature_names
bos.info()
bos.head()

```

```python
import matplotlib.pyplot as plt

#featurey
#
for i in range(1,14):
    plt.figure(figsize=(10,40))
    plt.subplot(13,1,i)
    plt.plot(boston.target,bos.iloc[:,i-1],'*')
    plt.title(boston.feature_names[i-1])
```

```python
#
from sklearn.linear_model import ridge_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y = train_test_split(bos,boston.target,test_size=0.3)


lm = LinearRegression()
#;
```

```python
scores = []

for i in range(1,14):
    lm.fit(train_X.iloc[:,i-1:i],train_y)
    score =lm.score(test_X.iloc[:,i-1:i],test_y)
    scores.append(score)

plt.plot(boston.feature_names,scores)
plt.show()
```
