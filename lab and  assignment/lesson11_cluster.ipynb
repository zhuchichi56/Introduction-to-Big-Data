```python
n_samples=1500
from sklearn import datasets,cluster
blobs= datasets.make_blobs(n_samples=n_samples)
noisy_circles =datasets.make_circles(n_samples=n_samples,noise=0.05)
noisy_moons=datasets.make_moons(n_samples=n_samples,noise=0.05)
no_structure=np.random.rand(n_samples,2),None
```

```python
from sklearn.preprocessing import StandardScaler
plt.figure(figsize=(8,8))
plt.subplots_adjust(
    left=0.02,right=0.92,bottom=0.001,top=0.95,wspace=0.05,hspace=0.01
)
dbscan=cluster.DBSCAN()
dataset=[
    (noisy_circles,{'eps':0.3,'min_samples':5,'n_clusters':2}),
    (noisy_moons,{'eps':0.3,'min_samples':5,'n_clusters':2}),
    (blobs,{'eps':0.3,'min_samples':5,'n_clusters':3}),
    (no_structure,{'eps':0.3,'min_samples':5,'n_clusters':3})
]


```

```python
number=0
for id,(dataset,params) in enumerate(dataset):
    X,y=dataset
    X=StandardScaler().fit_transform(X)
    two_means=cluster.MiniBatchKMeans(n_clusters=params['n_clusters'])
    dbscan= cluster.DBSCAN(eps=params['eps'],min_samples=params['min_samples'])
    
    models=(
        ('MiniBatchKMeans',two_means),
        ('DBSCAN',dbscan)
    )

    for name,algorithm in models:
        number=number+1
        algorithm.fit(X)
        y_pred=algorithm.labels_

        plt.subplot(len(dataset)*2,len(models)*2,number)
        plt.scatter(X[:,0],X[:,1],c=y_pred,s=10)
plt.show()
```
