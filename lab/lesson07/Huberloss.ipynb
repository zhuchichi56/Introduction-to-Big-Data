```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
norms = sm.robust.norms
```

```python
help(norms)

#huber loss
r = np.linspace(-6,6,1000)
huber = norms.HuberT(t = 1.345)
weight = huber.weights(r)
turkey = norms.TukeyBiweight(c = 1)

plt.plot(r,weight)
plt.plot(r,turkey.weights(r))


```
