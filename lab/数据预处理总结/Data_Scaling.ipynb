```python
import pandas as pd
import numpy as np
import seaborn as sns


```

```python
df = pd.read_csv("teenager_sns.csv")
#年龄有异常值
sns.displot(df["age"])
```

```python
from sklearn.preprocessing import scale
from sklearn.preprocessing import minmax_scale
Z_score_scaling =scale(df["age"],with_mean=df['age'].mean(),with_std=df['age'].std())
Scaling_01 = minmax_scale(df["age"])
```

```python
sns.displot(Z_score_scaling)
sns.displot(Scaling_01)

```
