```python
import pylab as pl
x = np.linspace(0,10,10)
y= np.sin(x)
pl.plot(x,y)
```

```python

from scipy import interpolate


function_line = interpolate.interp1d(x,y)
newx = np.linspace(0,10,100)
newy = function_line(newx)
pl.plot(newx,newy)

function_line = interpolate.lagrange(x,y)
newx = np.linspace(0,10,100)
newy = function_line(newx)
pl.plot(newx,newy)



data_df_age.isna().value_counts()
data_df_age.fillna('unkown')
data_df_age.replace(np.nan,'unkown')
```
